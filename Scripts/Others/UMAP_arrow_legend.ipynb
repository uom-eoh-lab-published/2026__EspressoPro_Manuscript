```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(2, 2), facecolor='none')

# Draw arrows
ax.arrow(0, 0, 0, 0.5, head_width=0.05, head_length=0.05, fc='black', ec='black', lw=6)  # Up
ax.arrow(0, 0, 0.5, 0, head_width=0.05, head_length=0.05, fc='black', ec='black', lw=6)  # Right

# Labels
ax.text(-0.25, 0.3, "UMAP-2", ha='center', va='center', rotation=90, fontsize=12)
ax.text(0.25, -0.25, "UMAP-1", ha='center', va='center', fontsize=12)

# Style
ax.set_xlim(-0.2, 1.2)
ax.set_ylim(-0.2, 1.2)
ax.set_aspect('equal')
ax.axis('off')

plt.tight_layout()

# Show with transparent background
fig.patch.set_alpha(0.0)
ax.patch.set_alpha(0.0)

plt.show()

# Optional: save transparent PNG
# fig.savefig("umap_axes.png", transparent=True, dpi=300)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(2, 2), facecolor='none')

# Draw arrows
# Up (unchanged)
ax.arrow(0, 0, 0, 0.5, head_width=0.05, head_length=0.05,
         fc='black', ec='black', lw=6)

# Left (reversed x-axis direction: dx is negative now)
ax.arrow(0, 0, -0.5, 0, head_width=0.05, head_length=0.05,
         fc='black', ec='black', lw=6)

# Labels
# Vertical label stays
ax.text(0.25, 0.3, "UMAP-2", ha='center', va='center',
        rotation=90, fontsize=12)

# Horizontal label now placed to the left
ax.text(-0.25, -0.25, "UMAP-1", ha='center', va='center',
        fontsize=12)

# Style / limits: extend to the left now
ax.set_xlim(-1.2, 0.2)
ax.set_ylim(-0.2, 1.2)
ax.set_aspect('equal')
ax.axis('off')

plt.tight_layout()

# Transparent background
fig.patch.set_alpha(0.0)
ax.patch.set_alpha(0.0)

plt.show()

# Optional: save transparent PNG
# fig.savefig("umap_axes_left.png", transparent=True, dpi=300)
```
